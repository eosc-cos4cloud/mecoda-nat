```python
def get_obs_from_query(query):
    observations = []
    n = 1
    url = f'{api_url}/observations.json?q="{query}"&per_page=200&page={n}'
    page = requests.get(url)
    if len(requests.get(url + '1000').json()) < 200:
        while len(page.json()) == 200:
            observations.extend(page.json())
            n += 1
            url = f'{api_url}/observations.json?q="{query}"&per_page=200&page={n}'
            page = requests.get(url + f'{n}')
        
        observations.extend(page.json())
    else:
        for year in range(2000, datetime.datetime.now().year + 1):
            while len(page.json()) == 200:
                url = f'{api_url}/observations.json?q="{query}"&per_page=200&page={n}&year={year}'
                page = requests.get(url)
                observations.extend(page.json())
                n += 1
            
    #__import__("pdb").set_trace()
    
    return observations
```

```python
# Extraer todas las observaciones por rangos de fechas
import time
start = time.time()

# primeros años

observations = []

url = f"https://natusfera.gbif.es/observations.json?per_page=200&order_by=observed_on&order=asc&page=1"
page = requests.get(url)
observations.extend(page.json())
year_change = observations[-1]['observed_on'].split("-")[0]


# resto de años
for year in track(range(int(year_change), datetime.datetime.now().year + 1)):
    if len(requests.get(f'{API_URL}/observations.json?year={year}&per_page=200&page=100').json()) < 200:
        n = 1
        url = f"https://natusfera.gbif.es/observations.json?per_page=200&year={year}&page={n}"
        page = requests.get(url)

        while len(page.json()) == 200:
            observations.extend(page.json())
            n += 1
            url = f"https://natusfera.gbif.es/observations.json?per_page=200&year={year}&page={n}"
            page = requests.get(url)

        observations.extend(page.json())            
    else:
        for month in range(1, 13):
            n = 1
            url = f"https://natusfera.gbif.es/observations.json?on={year}-{month}&per_page=200&page={n}"
            page = requests.get(url)
            while len(page.json()) == 200:
                observations.extend(page.json())
                n += 1
                url = f"https://natusfera.gbif.es/observations.json?on={year}-{month}&per_page=200&page={n}"
                page = requests.get(url)
            observations.extend(page.json())
        

end = time.time()

print("The time of execution of above program is :", end-start)
print(len(observations))
```

```python
def get_obs_from_id(id):
    """Get information on a specific observation given an id"""
    url = f'{API_URL}/observations/{id}.json'
    page = requests.get(url, verify=False)
    
    return page.json()
```
